## 3: il ciclo for

Per disegnare i poligoni precedenti, abbiamo ripetuto alcuni comandi pi volte, facendo copia incolla. Nei linguaggi di programmazione, per ripetere pi volte una serie di istruzioni, si usano dei costrutti appositi chiamati "cicli".

Uno dei cicli pi semplici  il ciclo "for", che permette di ripetere la stessa istruzione un numero prederminato di volte. Ad esempio, il seguente codice genera un poligono con N lati (nell'esempio, un pentagono con N=5):

```python
# riporta bob nello stato iniziale
bob.home()
bob.clear()

# disegna un poligono con N lati
N = 5
angolo = 360/N
# dimensione del lato del poligono
lato = 100

# sintassi per il ciclo for (nota i ":" alla fine della riga)
for i in range(N):
    # il blocco di codice che  indentato fa parte del ciclo for
    bob.forward(lato)
    bob.left(angolo)
```

Il ciclo "for"  un primo esempio di struttura di controllo: un'istruzione del linguaggio di programmazione che altera la sequenza di operazioni che vengono eseguite dal programma.

"range(N)"  una funzione predefinita del python che genera una sequenza con N elementi: tutti i numeri interi da 0 a N-1.

"i"  la variabile indice, che assume in sequenza tutti i valori generati da "range".

Ad esempio:

```python
for i in range(10):
    print(i)
```

## 4: Ciclo for e funzioni

Disegnare un poligono regolare  un compito generale che potremmo voler eseguire pi volte. In questo caso,  utile raggruppare il codice in una funzione:

```python
# definizione della funzione (nota i ":" finali)
def poligono_regolare(N, lato):
    # le varibili N e lato sono gli argomenti della funzione
    # il codice intentato (che inizia con il carattere "tab") fa parte della funzione
    angolo = 360/N
    for i in range(N):
        # Nota che il codice all'interno del ciclo for  indentato rispetto a quello della funzione
        bob.forward(lato)
        bob.left(angolo)
```

Se esegui il codice precedente, bob non esegue nessun comando, perch la funzione viene dichiarata, ma non viene eseguita. Per disegnare un poligono, dopo aver dichiarato la funzione, devi eseguirla:

```python
bob.home()
bob.clear()

# esegue la funzione definita in precedenza
# disegna un pentagono di lato 50
poligono_regolare(5, 50)
```

Avere le istruzioni per disegnare un poligono in una funzione, ci permette di disegnare figure pi complesse.
